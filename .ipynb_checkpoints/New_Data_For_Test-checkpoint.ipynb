```python
import pandas as pd
import numpy as np
```

```python
from ucimlrepo import fetch_ucirepo
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17) 
```

```python
# data (as pandas dataframes) 
X = breast_cancer_wisconsin_diagnostic.data.features 
y = breast_cancer_wisconsin_diagnostic.data.targets 
```

```python
X.columns
```

```python
data = X.copy()
data_processed = pd.DataFrame()
feature_groups = ['radius', 'texture', 'perimeter', 'area', 'smoothness', 
                  'compactness', 'concavity', 'concave_points', 
                  'symmetry', 'fractal_dimension']

# Step 1: Calculate means and drop original columns
for group in feature_groups:
    cols = [col for col in data.columns if col.startswith(group)]
    if len(cols) > 1:
        data_processed[f'{group}_mean'] = data[cols].mean(axis=1)

# Step 2: Recompute the "SE" (standard deviation) and drop original columns
for group in feature_groups:
    cols = [col for col in data.columns if col.startswith(group)]
    if len(cols) > 1:
        data_processed[f'{group}_se'] = data[cols].std(axis=1)

# Step 3: Recompute the "worst" (maximum) and drop original columns
for group in feature_groups:
    cols = [col for col in data.columns if col.startswith(group)]
    if len(cols) > 1:
        data_processed[f'{group}_worst'] = data[cols].max(axis=1)

data_processed
```

```python
data_processed.columns
```

```python
# Define the list of columns you want in the final DataFrame
selected_columns = [
    'texture_mean', 'symmetry_mean', 'fractal_dimension_mean', 
    'texture_se', 'area_se', 'smoothness_se', 'concavity_se', 
    'concave points_se', 'symmetry_se', 'fractal_dimension_se', 
    'smoothness_worst', 'symmetry_worst', 'fractal_dimension_worst', 
    'diagnosis'
]

# Ensure the selected columns exist in the processed DataFrame
existing_columns = [col for col in selected_columns if col in data_processed.columns]
data_selected = data_processed[existing_columns]

data_selected
```

```python
data_selected['concave points_se'] = np.std([X['concave_points1'], X['concave_points2'], X['concave_points3']], axis=0) / np.sqrt(3)
```

### Scale Data

```python
from sklearn.preprocessing import StandardScaler

scaled_data = StandardScaler().fit_transform(data_selected)
scaled_df = pd.DataFrame(scaled_data, columns=data_selected.columns)
scaled_df
```

```python
scaled_df.columns
```

```python
scaled_df = scaled_df[['texture_mean', 'symmetry_mean', 'fractal_dimension_mean', 'texture_se',
                       'area_se', 'smoothness_se', 'concavity_se', 'concave points_se',
                       'symmetry_se', 'fractal_dimension_se', 'smoothness_worst',
                       'symmetry_worst', 'fractal_dimension_worst']]
scaled_df
```

```python
scaled_df['diagnosis'] = y
```

### Save to CVS

```python
scaled_df.to_csv("test_data.csv", index=False)
```
